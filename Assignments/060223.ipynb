```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('titanic.csv')
data.head()
```

```python
data.info()
```

#### Number of passengers in each class

```python
data.Pclass.value_counts()
```

### Single group single column

#### count of passengers who survived from each class

```python
pclass_survived = data.groupby(['Pclass'])['Survived']
```

```python
pclass_survived.sum()
```

#### count of passengers who travelled with sibling

```python
pclass_sibsp = data.groupby(['Pclass'])['SibSp']
```

```python
pclass_sibsp.sum()
```

### Multiple Columns

#### Total amount paid by each class who survived

```python
pclass_survived_fare = data.groupby(['Pclass'])['Survived','Fare']
```

```python
pclass_survived_fare.sum()
```

### Multiple groups

 #### Count of Females & Males who survived in each class                          

```python
pclass_sex_survived = data.groupby(['Pclass','Sex'])['Survived']
```

```python
pclass_sex_survived.sum()
```

```python
pclass_sex_survived = data.groupby(['Pclass','Sex'],as_index=False)['Survived']
```

```python
pclass_sex_survived.sum()
```

### Multiple Functions 

#### Mean & sum of pclass & Survived

```python
pclass_survived_mean_sum = data.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
```

```python
pclass_survived_mean_sum
```

## pd.cut method

#### Creating the bins for ages

```python
data['age_bins']=pd.cut(data.Age,bins=3,labels=('young','middle','old'))
```
