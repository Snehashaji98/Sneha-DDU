## Type Casting with pandas

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Car_sales.csv')
```

```python
data.head()
```

 ### Drop na

```python
data.dropna(inplace=True)
```

```python
data.info()
```

### Checking the contents in latest launch column

```python
data.Latest_Launch.unique()
```

```python
data['Latest_Launch']=pd.to_datetime(data.Latest_Launch)
```

```python
data.head(10)
```

```python
data.info()
```

```python
data['year']=pd.DatetimeIndex(data.Latest_Launch).year
```

```python
data['month']=pd.DatetimeIndex(data.Latest_Launch).month
```

```python
data['day']=pd.DatetimeIndex(data.Latest_Launch).day
```

```python
data.head(10)
```

### Convert the dtype of fuel efficiency into int

```python
data.Fuel_efficiency=data.Fuel_efficiency.astype(int64)
```

```python
data.Fuel_efficiency
```

```python
data.head()
```

### Converting the numerical column into categorical column
#### Lets convert the month num to their corresponding names

```python
data.month.replace({
    1:'January',
    2:'February',
    3:'March',
    4:'April',
    5:'may',
    6:'June',
    7:'July',
    8:'August',
    9:'September',
    10:'October',
    11:'November',
    12:'December'
},inplace=True)
```
