### Treating outliers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Loading a dataset from seaborn for demonstration

```python
data=sns.load_dataset('car_crashes')
```

## Outlier treatment
### Boxplot

```python
for i in data.columns:
    if data[i].dtype!="object":
        sns.boxplot(y=data[i],color="red")
        plt.show();
```

```python
data.ins_losses.plot.hist(figsize=(6,3),cmap='brg');
```

### Setting outlier limits:

```python
def outlier_limit(col):
    Q3,Q1=np.nanpercentile (col,[75,25])
    IQR = Q3-Q1
    UL=Q3+1.5*IQR
    LL=Q1-1.5*Q1
    return UL,LL
```

```python
for i in data.columns:
    if data[i].dtype!="object":
        UL,LL=outlier_limit(data[i])
        data[i] = np.where((data[i]>UL)|(data[i]<LL), np.nan, data[i])
```

### Lets plot the barplot again to see the difference

```python
for i in data.columns:
    if data[i].dtype!="object":
        sns.boxplot(y=data[i],color="Green")
        plt.show();
```

## We can see that the outlier are treated.
