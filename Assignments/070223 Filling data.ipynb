### ?  Handlig Null Value

```python
data=pd.read_csv('Windspeed.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

### Filling a common value to all missing data

####  Lets try filling 0 to all the missing data

```python
data.fillna(0)
```

### Add missing data to individual columns

```python
data.fillna({
    "Temperature":0,
    "Windspeed":5,
    "Status":"sunny"
})
```

### Forward Fill(ffill)row
##### fill the data from the row above the mising value

```python
data.fillna(method="ffill")
```

### Backward Fill(row)
##### Fill the missing data from the existing data below

```python
data.fillna(method="bfill")
```

### Limiting the forward/backward fill
##### We can limit the num.of rows or column getting filled

```python
data.fillna(method="ffill",limit=1)
```

###  Filling with pandas object
#### Fill the missing values with panadas object like (df.sum(),df.max(), etc

```python
data.fillna(data.mean())
```

```python
data.fillna(data.max())
```

### Filling for specific range of columns

```python
data.fillna(data.mean()['Temperature':'windspeed'])
```
