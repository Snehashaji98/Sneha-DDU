```python
df['Date']=pd.to_datetime(df.Date)
df.head(10)
```

### Extracting Year,Month and day from the date column

```python
df['Year']=pd.DatetimeIndex(df.Date).year
df['Month']=pd.DatetimeIndex(df.Date).month
df['Day']=pd.DatetimeIndex(df.Date).day
df.head(10)
```

### Droping Lat, Long and Date


```python
df.drop(columns=['Lat','Long','Date','Province/State'],inplace=True)
```

```python
df.head(10)
```

```python
df.Month.replace({
    1:"January",
    2:"February",
    3:"March",
    4:"April",
    5:"May",
    6:"June",
    7:"July",
    8:"August",
    9:"September",
    10:"October",
    11:"November",
    12:"December"
}, inplace=True)
df.sample()
```

```python
df.rename(columns={'Country/Region':'Region'}, inplace= True)
df.head(10)
```
