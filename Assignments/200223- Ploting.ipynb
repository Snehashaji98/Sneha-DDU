### Univariate analysis

### KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.PassengerId)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.Age);
```

```python
plt.figure()
sns.kdeplot(data=data.Fare);
```

```python
sns.histplot(data.Fare)
plt.show()
```

```python
sns.histplot(data.Age)
plt.show()
```

```python
sns.boxplot('Age',data=data)
plt.show()
```

```python
sns.boxplot('Fare',data=data)
plt.show()
```

### Multi variate analysis
 ### Line Plot

```python
plt.figure(figsize=(6,3))
sns.lineplot('Age','Fare',data=data)
plt.title("Age vs Fare")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.lineplot('Age','Fare',data=data, hue='Sex')
plt.title("Age vs Fare")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.lineplot('Age','Fare',data=data, hue='Pclass')
plt.title("Age vs Fare")
plt.show()
```

```python
sns.lineplot('Age','Survived',data=data)
plt.title("Age vs Survived")
plt.show()
```

```python
sns.lineplot('Survived','Age', data=data, hue='Sex')
plt.title("Age vs Survived")
plt.show()
```

### PIE CHART

```python
data.columns
```

```python
Pclass_Survived = data.groupby(['Pclass'])['Survived'].sum()
```

```python
Pclass_Survived
```

```python
sns.set_style('ticks')
Pclass_Survived.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_sex_Survived=data.groupby(['Pclass','Sex'])['Survived'].sum()
```

```python
Pclass_sex_Survived.reset_index()
```

### Pie Chart

```python
Pclass_sex_Survived.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left', borderaxespad=5)
plt.show()
```

### Bar Chart

```python
plt.figure(figsize=(4,3))
sns.countplot('Sex',hue='Survived',data=data)
plt.show()
```

### Count Plot

```python
plt.figure(figsize=(4,2))
sns.countplot('Survived',data=data);
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Survived',data=data)
plt.title('Pclass:Survived vs Dead')
plt.show()           
```

### Factor plot

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived',hue='Sex',data=data)
plt.show()
```

### Violin Plot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(5,4))
sns.violinplot("Sex","Age",hue="Survived",data=data)
plt.show()
```

### Violin plot

```python
sns.set_style('whitegrid')
plt.figure(figsize=(6,4))
sns.violinplot("Survived","Age",hue="Survived",data=data)
plt.show()
```

### Bar plot

```python
plt.figure(figsize=(8,4))
sns.barplot('SibSp', 'Survived', data=data)
plt.title("SibSp &Survived")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot('Parch', 'Survived', data=data)
plt.title("Parch & Survived")
plt.show()
```
